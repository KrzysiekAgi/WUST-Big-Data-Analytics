# Lotka-Voltera Model

### Parameters

```python
time = 1000
delta_t = 1/(time/10)
rabbits = np.zeros(time + 1)
foxes = np.zeros(time + 1)
alpha = 0.01

rabbits[0] = 300
foxes[0] = 150
```

### Model

```python
def lotka_voltera(r0, f0, alpha, delta_t):
    drdt = 2*r0 - alpha*r0*f0
    dfdt = -f0 + alpha*r0*f0
    r1 = r0 + drdt*delta_t
    f1 = f0 + dfdt*delta_t
    return r1, f1
```

### Calculation

```python
for t in range(time):
    rabbits[t+1], foxes[t+1] = lotka_voltera(rabbits[t], foxes[t], alpha, delta_t)
    
last_date = t+1
X = np.zeros(time + 1)
for t in range(time + 1):
    X[t] = t
    
plt.title("Lotka Voltera Model")
plt.xlabel("time")
plt.ylabel("poplation")
plt.plot(X, rabbits, label="Rabbits")
plt.plot(X, foxes, label = "Foxes")
plt.xlim(0, last_date)
plt.legend()
plt.show()
```
