```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df1 = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
df1.index.freq = 'MS'
df1.dropna(inplace=True)
```

```python
df1.head()
```

```python
df1['Thousands of Passengers'].plot()
```

```python
d1 = df1['Thousands of Passengers'].diff(1)
```

```python
df1['d1'] = d1
df1['d1'].plot()
```

```python
d1_12 = d1.diff(12)
df1['d1_12'] = d1_12
df1['d1_12'].plot()
```

```python
d1_log = np.log(df1['Thousands of Passengers']).diff(1)
df1['d1_log'] = d1_log
df1['d1_log'].plot()
```
