# MA model for Returns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.arima_model import ARMA
from scipy.stats.distributions import chi2
import statsmodels.tsa.stattools as sts
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set()
```

```python
def LLR_test2(L1, L2, DF=1):
    LR = 2*(L2-L1)
    return chi2.sf(LR,DF).round(3)
```

```python
raw_csv_data = pd.read_csv("Index2018.csv")
df_comp = raw_csv_data.copy()
df_comp.date = pd.to_datetime(df_comp.date, dayfirst=True)
df_comp.set_index("date", inplace=True)
df_comp = df_comp.asfreq('b') #business days
df_comp = df_comp.fillna(method='ffill')
```

```python
df_comp['market_value'] = df_comp.ftse
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
```

```python
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size],df_comp.iloc[size:]
```

```python
df['returns'] = df.market_value.pct_change(1).mul(100)
df = df.iloc[1:]
```

```python
df.returns.plot()
plt.title("FTSE returns", size=20)
plt.show()
```

```python
sgt.plot_acf(df.returns, zero=False, lags=40)
plt.title("ACF for returns", size=24)
plt.show()
```

```python
model_ret_ma_1 = ARMA(df.returns, order=(0,1))
results_ret_ma_1 = model_ret_ma_1.fit()
results_ret_ma_1.summary()
```

```python
# Higher order model
model_ret_ma_2 = ARMA(df.returns, order=(0,2))
results_ret_ma_2 = model_ret_ma_2.fit()
results_ret_ma_2.summary()
```

```python
LLR_test2(results_ret_ma_1.llf, results_ret_ma_2.llf, 1)
```

```python
model_ret_ma_6 = ARMA(df.returns, order=(0,6))
results_ret_ma_6 = model_ret_ma_6.fit()
results_ret_ma_6.summary()
```

```python
model_ret_ma_8 = ARMA(df.returns, order=(0,8))
results_ret_ma_8 = model_ret_ma_8.fit()
results_ret_ma_8.summary()
```

```python
df['res_ret_ma_8'] = results_ret_ma_8.resid
```

```python
df.res_ret_ma_8.plot(figsize=(20,5))
plt.title("Residuals for Ma returns", size=24)
plt.show()
```

```python
np.sqrt(df.res_ret_ma_8.var()) #standard deviation
```

```python
sts.adfuller(df.res_ret_ma_8)
```

```python
sgt.plot_acf(df.res_ret_ma_8, zero=False, lags=40)
plt.title("ACF for residuals", size=24)
plt.show()
```
