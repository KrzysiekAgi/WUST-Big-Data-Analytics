31. Let S={(1,1),(1,4),(1.1,2),(1,1.1),(1.2,2.2),(5,1)(−1,−1),(−1,−4),(−5.1,−2.1),(−5.2,−0.9),(−5.1,−1.1),(−5,−1)(−6,2),(−6.2,2.1),(−5,3),(−6,3.1),(−6.2,3.2),(−5.5,3.3)}. Use k- menas algorithm for clustering S. Return pictures of clusters for k= 2,3,4,5

```python
data = {'x': [1,1,1.1,1,1.2,5,-1,-1,-5.1,-5.2,-5.1,-5,-6,-6.2,-5,-6,-6,-5.5],'y': [1,4,2,1.1,2.2,1,-1,-4,-2.1,-0.9,-1.1,-1,2,2.1,3,3.1,3.2,3.3]}
df = DataFrame(data,columns=['x','y'])
numberOfCluster=[2,3,4,5]

for number in numberOfCluster:
    kmeans = KMeans(n_clusters=number).fit(df)
    centroids = kmeans.cluster_centers_

    plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
    plt.title("K means clustering algorithm for " + str(number) + " clusters")
    plt.show()
```

32. (use NLTK). Let S1, S2, S3 be the sets of all words with at most 7 letters from text1,text2, text3, respectively. Let S=S1∪S2∪S3. Use any reasonable algorithm for clustering S (with edit distance). Return sizes of clusters for k= 2,3,10 clusters

```python
from nltk.book import *
```

```python
def chooseWords(S, size):
    wordsToConsider = set()
    for word in S:
        if len(word)<=size:
            wordsToConsider.add(word)
    return wordsToConsider
         
S1 = chooseWords(set(text1), 7)
S2 = chooseWords(set(text2), 7)
S3 = chooseWords(set(text3), 7)
S = S1 | S2 | S3
lst=list()

for word in S:
    dist = nltk.edit_distance(word, "what?")
    t = (dist, len(word))
    lst.append(t)
    
df = DataFrame(lst, columns=['edit distance', 'length'])
print(df)

numberOfCluster=[2,3,10]

for number in numberOfCluster:
    kmeans = KMeans(n_clusters=number).fit(df)
    centroids = kmeans.cluster_centers_

    plt.scatter(df['length'], df['edit distance'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
    plt.title("K means clustering algorithm for " + str(number) + " clusters")
    plt.show()
```
