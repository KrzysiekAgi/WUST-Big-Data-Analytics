```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
eq1 = x**2 + 4*x - 5
solve(eq1, x)
```

```python
x, y = symbols('x y')
polyn = x**3 + 7*x*y + y**2
a = Poly(polyn, x, y)
a.coeffs()
```

```python
x, y, a = symbols('x y a')
expr = 2*x**3 + x**4 + a*x**3
collect(expr, [x, y])
```

```python
x, y = symbols('x y')
eq = x**2*y
diff(eq, x) # first derivative
diff(diff(eq, x), x) # second derivative
diff(eq, x, x) # same as above
```
