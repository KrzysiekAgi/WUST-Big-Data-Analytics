# 2.1 Linear harmonic oscilator

Harmonic osiclator equation

```python
x = Function("x")
t, β, ω, Ω, F = symbols("t β ω Ω F")
Eq(Derivative(x(t), t, t) + 2*β*Derivative(x(t), t) + ω*ω*x(t), F*cos(Ω*t))
```

### 2.1.1 Case β = 0, ω = 1, F = 0, Ω = 0

```python
x = Function("x")
t = symbols("t")
difeq1 = diff(x(t),t,t) + x(t)

y = dsolve(difeq1,x(t)) 
C = solve([y.rhs.subs(t,0) - 1.0, y.rhs.diff(t).subs(t,0) - 0.0])
y = y.subs(C)
y = lambdify(t, y.rhs, "numpy")

t = np.arange(0,20,0.1)
plt.plot(t,y(t))
difeq1
```

### 2.1.2 Case β = 1/2, ω = 1, F = 0, Ω = 0

```python
x = Function("x")
t = symbols("t")

# solve symbolically
difeq2 = Eq(Derivative(x(t), t, t) + Derivative(x(t), t) + x(t))
y = dsolve(difeq2, x(t)) 

# calculate C1 and C2
C = solve([y.rhs.subs(t,0) - 1.0, y.rhs.diff(t).subs(t,0) - 0.0])
y = y.subs(C)
y = lambdify(t, y.rhs, "numpy")

t = np.arange(0,20,0.1)
plt.plot(t,y(t))
```

### 2.1.3 Case β = 1/2, ω = 1, F = 1, Ω = 2

```python
x = Function("x")
t = symbols("t")

#solve symbolically
difeq3 = Eq(Derivative(x(t), t, t) + Derivative(x(t), t) + x(t), cos(2*t))
y = dsolve(difeq3, x(t))

# calculate C1 and C2
# initial conditions: y(0)=1 and y'(0)=0
C = solve([y.rhs.subs(t,0) - 1.0, y.rhs.diff(t).subs(t,0) - 0.0]) 
y = y.subs(C)
y = lambdify(t, y.rhs, "numpy")

t = np.arange(0,20,0.1)
plt.plot(t,y(t))
```

### 2.1.4 Case β = 0, ω = 1, F = 1, Ω = 1

```python
x = Function("x")
t = symbols("t")

#solve symbolically
difeq3 = Eq(Derivative(x(t), t, t) + x(t), cos(t))
y = dsolve(difeq3, x(t))

# calculate C1 and C2
C = solve([y.rhs.subs(t,0) - 1.0, y.rhs.diff(t).subs(t,0) - 0.0])
y = y.subs(C)
y = lambdify(t, y.rhs, "numpy")

t = np.arange(0,100,0.1)
plt.plot(t,y(t))
```
