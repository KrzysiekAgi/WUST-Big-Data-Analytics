# 1.1 Four stationary points

```python
from sympy import Symbol
x,y = symbols('x y')
P = 16*x**2 + 9*y**2 - 25
Q = 16*x**2 - 16*y**2
solve({P,Q},{x,y})
```

```python
a = diff(P, x)
b = diff(P, y)
c = diff(Q, x)
d = diff(Q, y)
print(a, b, c, d)
```

```python
x1 = -1
y1 = 1
a1 = a.subs(x,x1)
b1 = b.subs(y,y1)
c1 = c.subs(x,x1)
d1 = d.subs(y,y1)
q = a1*d1-b1*c1
p = -a1-d1
r = p**2-4*q
print(p,q,r)
```
